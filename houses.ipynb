```python
import googlemaps
import pandas as pd
```

```python
houses = pd.read_csv('/content/drive/MyDrive/경기도 안양시_공동주택 현황_20231212.csv', encoding='utf-8')
houses.head()
```

```python
gmaps = googlemaps.Client(key='AIzaSyD2TwhZuJjii-NP8_ZD4P9hG18lWaRrcbs')
```

```python
lat = []
lng = []

for address in houses['새주소']:
  try:
    address = '안양시 ' + address
    tmpMap = gmaps.geocode(address, language='ko')
    tmpLoc = tmpMap[0].get('geometry')
    lat.append(tmpLoc['location']['lat'])
    lng.append(tmpLoc['location']['lng'])
  except:
    print(f"{address}: Address Error")
    lat.append(None)
    lng.append(None)
```

```python
houses['longtitude'] = lng
houses['latitude'] = lat
```

```python
houses.head()
```

```python
import folium
```

```python
map_folium = folium.Map(location=[37.3942527, 126.9568209], zoom_start = 14, )
```

```python
map_folium
```

```python
for _, row in houses.iterrows():
  popup = folium.Popup(row['아파트명'], max_width=200)
  folium.Circle(location=[row['latitude'], row['longtitude']], popup = popup).add_to(map_folium)
```
